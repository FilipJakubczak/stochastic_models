```python
import numpy as np
import math
import csv
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 22})
```

# Functions

```python
def langevin_reconstruction(data, bin_number, time_lag, dt):
    '''Performs langevin reconstruction.'''
    n = len(data)
    tab_d = np.zeros((bin_number, 6))
    data = np.array(data)
    max_x = np.amax(data)
    min_x = np.amin(data)
    range_x = np.absolute(max_x - min_x)
    bin_width = range_x / bin_number
    data_x = np.zeros(bin_number)
    for i in range(bin_number):
        data_x[i] = min_x + (2 * (i + 1) - 1) * 0.5 * bin_width
    prob_den_x2x1 = np.zeros((bin_number, bin_number))
    sum_x1 = np.zeros(bin_number)
    for i in range(n - time_lag):
        x1 = np.floor((data[i]-min_x)/bin_width)+1
        if x1 == (bin_number + 1):
            x1 = bin_number
        x2 = np.floor((data[i + time_lag] - min_x) / bin_width) + 1
        if x2 == (bin_number + 1):
            x2 = bin_number
        prob_den_x2x1[int(x1) - 1][int(x2) - 1] = prob_den_x2x1[int(x1) - 1][int(x2) - 1] + 1
        sum_x1[int(x1) - 1] = sum_x1[int(x1) - 1] + 1;
    for i in range(bin_number):
        if sum_x1[i] > 0:
            for j in range(bin_number):
                prob_den_x2x1[i][j] = prob_den_x2x1[i][j] / sum_x1[i] / bin_width


    for i in range(bin_number):
        for j in range(bin_number):
            tab_d[i,1] = tab_d[i,1] + (data_x[i] - data_x[j]) * prob_den_x2x1[j,i] * bin_width
            tab_d[i,2] = tab_d[i,2] + ((data_x[i] - data_x[j]) ** 2) * prob_den_x2x1[j,i] * bin_width;
            tab_d[i,3] = tab_d[i,3] + ((data_x[i] - data_x[j]) ** 3) * prob_den_x2x1[j,i] * bin_width
            tab_d[i,4] = tab_d[i,4] + ((data_x[i] - data_x[j]) ** 4) * prob_den_x2x1[j,i] * bin_width;

    tab_d[:,0] = data_x[:]
    tab_d[:,1] = tab_d[:,1] / [time_lag * dt]
    tab_d[:,2] = tab_d[:,2] / (2 * time_lag * dt)
    tab_d[:,3] = tab_d[:,3] / (3 * time_lag * dt)
    tab_d[:,4] = tab_d[:,4] / (4 * time_lag * dt)
    tab_d[:,5] = sum_x1[:]
    return tab_d

def plot(x, y, xlabel: str = '', ylabel: str = '', title: str = ''):
    fig = plt.figure(figsize=(15,5))
    plt.plot(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
    
```

```python
tau = 0.01
x0 = 0.1
n = 100000
x = [x0]
```

```python
for i in range(n - 1):
    random = np.random.normal(0, 1)
    x.append(x[i] + (x[i] - x[i] ** 3) * tau + math.sqrt(4 * tau) * random)
```

```python
plot(x=range(n), y=x, xlabel='index', ylabel='x', title=r'$\frac{dX}{dt}=X-X^3+2\Gamma(t)$')
```

```python
bin_number = 45
dt = 0.01
time_lag = 1
tab_d = langevin_reconstruction(x, bin_number, time_lag, dt)
```

```python
plot(x=tab_d[:,0], y=tab_d[:, 1], xlabel='x', ylabel='D(1)')
```

```python
plot(x=tab_d[:,0], y=tab_d[:, 2], xlabel='x', ylabel='D(2)')
```
